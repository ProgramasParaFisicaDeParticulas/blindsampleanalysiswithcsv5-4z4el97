```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

particles = pd.read_csv('https://raw.githubusercontent.com/ProgramasParaFisicaDeParticulas/blindsampleanalysiswithcsv5-4z4el97/main/Sample5.csv', delimiter=',')
particles.head()
```

```python
plt.hist(particles['Q'], bins=3, color='blue', alpha=0.7, range= (-1,1))
plt.xlabel('Carga (GeV)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Carga')
plt.show()

cantidad_valores_cero = (particles['Q'] == 0).sum()
print("La cantidad de datos con valor 0 en la columna 'Q' es:", cantidad_valores_cero)
```

```python
promedio_iso = particles['iso'].mean()
print("El promedio de la columna 'iso' es:", promedio_iso)
desviacion_estandar_iso = particles['iso'].std()
print("La desviación estándar de la columna 'iso' es:", desviacion_estandar_iso)
valor_max_iso = particles['iso'].max()
print("El valor máximo de la columna 'iso' es:", valor_max_iso)

plt.hist(particles['iso'], bins=5, color='blue', alpha=0.7, range = (0,25000))
plt.xlabel('iso')
plt.ylabel('Frecuencia')
plt.title('Histograma de iso')
plt.show()
```

```python
particles_filtrado = particles[(particles['iso'] >= 0) & (particles['iso'] <= 1)]
cantidad_datos_iso = particles_filtrado.shape[0]
print("Cantidad de datos después del filtro:", cantidad_datos_iso)
particles_filtrado.head()
```

```python
plt.hist(particles_filtrado['iso'], bins=5, color='blue', alpha=0.7)
plt.xlabel('iso')
plt.ylabel('Frecuencia')
plt.title('Histograma de iso')
plt.show()
```

```python
promedio_pt = particles_filtrado['pt'].mean()
print("El promedio de la columna 'pt' es:", promedio_pt)
desviacion_estandar_pt = particles_filtrado['pt'].std()
print("La desviación estándar de la columna 'pt' es:", desviacion_estandar_pt)
valor_max_pt = particles_filtrado['pt'].max()
print("El valor máximo de la columna 'pt' es:", valor_max_pt)
valor_min_pt = particles_filtrado['pt'].min()
print("El valor mínimo de la columna 'pt' es:", valor_min_pt)

plt.hist(particles_filtrado['pt'], bins=120, color='blue', alpha=0.7)
plt.xlabel('Momento transversal (GeV)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Momeonto Transversal iso[0,30]')
plt.grid(True)
plt.show()
```
